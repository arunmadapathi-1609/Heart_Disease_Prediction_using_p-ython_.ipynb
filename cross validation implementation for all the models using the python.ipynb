```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
heart_data=pd.read_csv('/content/heart_disease_data.csv')
```

```python
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

#Train Test Split


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
# comparing the perfomance of the model

models=LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()
```

```python
def compare_models_train_test():
  for model in models:
    model.fit(x_train,y_train)
    test_data_prediction=model.predict(x_test)
    accuracy=accuracy_score(y_test,test_data_prediction)
    print('Accuracy score of the model ', model , '=', accuracy)

```

```python
compare_models_train_test()
```

#Cross Validation


```python
# using the logistic Regression

cv_score_lr=cross_val_score(LogisticRegression(max_iter=1000), x,y,cv=5) # cv=5 represents the no of flods
```

```python
print(cv_score_lr)
```

```python
mean_accuracy_lr=sum(cv_score_lr)/len(cv_score_lr) # sum of accuracy/no of iterations
mean_accuracy_lr=mean_accuracy_lr*100
mean_accuracy_lr=round(mean_accuracy_lr,2)
print(mean_accuracy_lr)
```

```python
# Support vector classsifier
cv_score_SVC=cross_val_score(SVC(kernel='linear'), x,y,cv=5) # cv=5 represents the no of flods

print(cv_score_SVC)
mean_accuracy_SVC=sum(cv_score_SVC)/len(cv_score_SVC) # sum of accuracy/no of iterations
mean_accuracy_SVC=mean_accuracy_SVC*100
mean_accuracy_SVC=round(mean_accuracy_SVC,2)
print(mean_accuracy_SVC)

```

```python
# creating afunction to compare the models
models=LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()
```

```python
def compare_models_cross_validation():
  for model in models:
    cv_score = cross_val_score(model ,x, y, cv=5)
    mean_accuracy=sum(cv_score)/len(cv_score) # sum of accuracy/no of iterations
    mean_accuracy=mean_accuracy*100
    mean_accuracy=round(mean_accuracy,2)
    print('cross validation accuraciesfor the ', model,'=',cv_score)
    print('accuracy % of the ', model,mean_accuracy)
    print('----------------------------------------------')
```

```python
compare_models_cross_validation()
```
